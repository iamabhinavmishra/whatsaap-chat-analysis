```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with CI Equity Learning (1).txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern,data)[1:]
messages
```

```python
dates = re.findall(pattern,data)
dates
```

```python
df = pd.DataFrame({'user_message':messages, 'message_dates': dates})
df['message_dates'] = pd.to_datetime(df['message_dates'], format='%d/%m/%y, %H:%M - ')
df.rename(columns={'message_dates': 'date'}, inplace=True)
```

```python
#seprate user and message
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users 
df['message'] = messages 
df.drop(columns=['user_message'], inplace=True)
df.head()
```

```python
df['year'] = df['date'].dt.year
```

```python
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df.head()
```
